```python
df = pd.read_csv('/kaggle/input/datasets/praveengovi/emotions-dataset-for-nlp/train.txt',sep = ';',header = None,names = ['text','emotion'])
```

```python
unique_emonations = df['emotion'].unique()
```

```python
emotions_number ={}
i =0
for emotion in unique_emonations:
  emotions_number[emotion] =i
  i+=1

df['emotion'] = df['emotion'].map(emotions_number)
```

```python
df['emotion'].unique()
```

```python
df.head()
```

```python
df['text'] = df['text'].apply(lambda x:x.lower())
```

```python
import string
def remove_punc(txt):
  return txt.translate(str.maketrans('','',string.punctuation))
```

```python
df['text'] = df['text'].apply(remove_punc)
```

```python
def remove_numbers(txt):
    new = ""
    for i in txt:
        if not i.isdigit():
            new = new + i
    return new

df['text'] = df['text'].apply(remove_numbers)
```

```python
def remove_emojis(txt):
    new = ""
    for i in txt:
        if i.isascii():
            new += i
    return new

df['text'] = df['text'].apply(remove_emojis)
```

```python
import nltk
```

```python

from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
```

```python
nltk.download('punkt')
nltk.download('stopwords')
```

```python
stop_words = set(stopwords.words('english'))
```

```python
df.loc[0]['text']
```

```python
def remove(txt):
  words = txt.split()
  cleaned = []
  for i in words:
    if not i in stop_words:
      cleaned.append(i)

  return ' '.join(cleaned)
```

```python
df['text'] = df['text'].apply(remove)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(df['text'], df['emotion'], test_size=0.20, random_state=42)
```

```python
from sklearn.feature_extraction.text import CountVectorizer, TfidfVectorizer

from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score

bow_vectorizer = CountVectorizer()
X_train_bow = bow_vectorizer.fit_transform(X_train)
X_test_bow = bow_vectorizer.transform(X_test)


nb_model = MultinomialNB()
nb_model.fit(X_train_bow, y_train)


pred_bow = nb_model.predict(X_test_bow)
print(accuracy_score(y_test, pred_bow))
```

```python
pred_bow
```

```python
tfidf_vectorizer = TfidfVectorizer()
X_train_tfidf = tfidf_vectorizer.fit_transform(X_train)
X_test_tfidf = tfidf_vectorizer.transform(X_test)


nb2_model = MultinomialNB()
nb2_model.fit(X_train_tfidf,y_train)
```

```python
y_pred = nb2_model.predict(X_test_tfidf)
```

```python
print(accuracy_score(y_test, y_pred))
```

```python
from sklearn.linear_model import LogisticRegression

logistic_model = LogisticRegression(max_iter=1000)
logistic_model.fit(X_train_tfidf,y_train)
```

```python
log_pred = logistic_model.predict(X_test_tfidf)
print(accuracy_score(y_test,log_pred ))
```
